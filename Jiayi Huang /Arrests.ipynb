```python
import pandas as pd
import numpy as np

incidents = pd.read_csv("Arrests.csv") #reading the data
Incidents = incidents['INCIDENTNEIGHBORHOOD'].value_counts() #counting the number of times each neighborhood was listed
Incidents.sort_values() #sorting from lowest number of incidents to highest
```

```python
incidents['INCIDENTNEIGHBORHOOD'].value_counts().tail(15).plot(kind='bar') #creating a bar graph to compare the number of incidentsfor each neighborhood
```
